```python
# Load module

from transformers import AutoProcessor, MusicgenForConditionalGeneration

processor = AutoProcessor.from_pretrained("facebook/musicgen-small")
model = MusicgenForConditionalGeneration.from_pretrained("facebook/musicgen-small")
```

```python
# Text-Conditional Generation
import scipy
import random
import string
import time
from IPython.display import Audio

def play_sound(file_path):
    display(Audio(file_path, autoplay=True))


start_time = time.time()

inputs = processor(
    text=["Ancient Chinese style music","Beijing Opera Style Chinese pop music"],
    padding=True,
    return_tensors="pt",
)
audio_values = model.generate(**inputs, do_sample=True, guidance_scale=3, max_new_tokens=512)
end_time = time.time()
print(f"模型生成用时: {int(end_time - start_time)} 秒")  # 计算并打印用时



# 生成一个长度为5的随机字符串
random_str = ''.join(random.choices(string.ascii_letters + string.digits, k=5))

# 修改文件名，加入随机字符串
file_name = f"audio_output/musicgen_out_{random_str}.wav"

# 使用修改后的文件名保存文件
sampling_rate = model.config.audio_encoder.sampling_rate
scipy.io.wavfile.write(file_name, rate=sampling_rate, data=audio_values[0, 0].numpy())

# Listen in Jupyter
sampling_rate = model.config.audio_encoder.sampling_rate
Audio(audio_values[0].numpy(), rate=sampling_rate, autoplay=True)
```
